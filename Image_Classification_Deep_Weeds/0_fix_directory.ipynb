# Train test split

```python
# split dataset
df_splitted_no_negative = df[df['Species'] != 'Negative']
df_splitted_negative = df[df['Species'] == 'Negative']
print(len(df_splitted_negative),
len(df_splitted_no_negative),
len(df_splitted_no_negative)/8)
df_splitted_negative_sampeled = df_splitted_negative.sample(1050)
len(df_splitted_negative_sampeled)
df = df_splitted_no_negative.append(df_splitted_negative_sampeled)
```

```python
# keep the labels in the training data to reorder the directorys 
# https://machinelearningmastery.com/how-to-load-large-datasets-from-directories-for-deep-learning-with-keras/

# data/
# data/train/
# data/test/red/
# data/test/blue/
# data/test/
# data/test/red/
# data/test/blue/
# data/validation/
# data/validation/red/
# data/validation/blue/
```

```python
# 80% (Train) , 10% (Validatino),  10% (Test)

X = df
y = df['Species']
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.8)

test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5)
```

```python
X_train['origin'] = "train"
X_valid['origin'] = "valid"
X_test['origin'] = "test"
```

```python
df_splitted = X_train
df_splitted = df_splitted.append(X_valid)
df_splitted = df_splitted.append(X_test)
```
