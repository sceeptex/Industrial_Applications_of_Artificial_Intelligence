```python
import torch
from transformers import XLNetTokenizer, XLNetForSequenceClassification
```

```python
tokenizer = XLNetTokenizer.from_pretrained("xlnet-base-cased")
```

```python
model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased")
```

```python
inputs = tokenizer("Hello, my dog not is cute", return_tensors="pt")
```

```python
with torch.no_grad():
    logits = model(**inputs).logits
```

```python
predicted_class_id = logits.argmax().item()
```

```python
model.config.id2label[predicted_class_id]
```

### use pretrained **labels**

```python
len(model.config.id2label)
```

```python
num_labels = len(model.config.id2label)
model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased", num_labels=num_labels)

```

```python
labels = torch.tensor(1)
loss = model(**inputs, labels=labels).loss
round(loss.item(), 2)
```

## Example of multi-label classification

```python
import torch
from transformers import XLNetTokenizer, XLNetForSequenceClassification

tokenizer = XLNetTokenizer.from_pretrained("xlnet-base-cased")
model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased", problem_type="multi_label_classification")




```

```python
model.config.id2label
```

```python
inputs = tokenizer("I love my dog", return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```

```python
num_labels = len(model.config.id2label)
model = XLNetForSequenceClassification.from_pretrained("xlnet-base-cased", num_labels=num_labels)

num_labels = len(model.config.id2label)
labels = torch.nn.functional.one_hot(torch.tensor([predicted_class_id]), num_classes=num_labels).to(
    torch.float
)
loss = model(**inputs, labels=labels).loss
loss.backward()
```

```python
print(f'Keys       : {list(text_preprocessed.keys())}')
print(f'Shape      : {text_preprocessed["input_word_ids"].shape}')
print(f'Word Ids   : {text_preprocessed["input_word_ids"][0, :12]}')
print(f'Input Mask : {text_preprocessed["input_mask"][0, :12]}')
print(f'Type Ids   : {text_preprocessed["input_type_ids"][0, :12]}')
```

Using the BERT model

```python
bert_model = hub.KerasLayer(tfhub_handle_encoder)
```

```python
bert_results = bert_model(text_preprocessed)

print(f'Loaded BERT: {tfhub_handle_encoder}')

print(f'Pooled Outputs Shape:{bert_results["pooled_output"].shape}') 
print(f'Pooled Outputs Values:{bert_results["pooled_output"][0, :12]}') #sequence embeddings

print(f'Sequence Outputs Shape:{bert_results["sequence_output"].shape}')
print(f'Sequence Outputs Values:{bert_results["sequence_output"][0, :12]}')
```

- **pooled_output** represents each input sequence as a whole. The shape is [batch_size, H]. You can think of this as an **embedding for the entire movie** review.
- **sequence_output** represents each input token in the context. The shape is [batch_size, seq_length, H]. You can think of this as a contextual embedding for every token in the movie review.
- **encoder_outputs** are the intermediate activations of the L Transformer blocks. outputs["encoder_outputs"][i] is a Tensor of shape [batch_size, seq_length, 1024] with the outputs of the i-th Transformer block, for 0 <= i < L. The last value of the list is equal to sequence_output.

```python
def useBertAsWordEmbeddings(sentece, printEmbeddings = False, showPloty = True):
  text_test = [sentece]
  text_preprocessed = bert_preprocess_model(text_test)
  bert_results = bert_model(text_preprocessed)
  if printEmbeddings:
    print(f'Pooled Outputs Values:{bert_results["pooled_output"]}')
  if showPloty: 

    fig = px.bar(y = bert_results["pooled_output"][0])
    fig.show()
```

```python
useBertAsWordEmbeddings('I love you')
```

```python
useBertAsWordEmbeddings('I hate you')
```
